```python

class OneMRecipeDataset(Dataset):
    def __init__(self, partition):
        if partition == 'train':
            f1 = open('imgs_train.json')
        if partition == 'test':
            f1 = open('imgs_test.json')
        if partition == 'val':
            f1 = open('imgs_val.json')

        self.partition = partition
        self.imgs = json.load(f1)

        #APPLY TRANSFORMS
        if self.partition == 'train':
            xform = transforms.Compose([
                transforms.RandomResizedCrop(224),
                transforms.RandomHorizontalFlip(),
                transforms.ToTensor()])
        if self.partition == 'test' or self.partition == 'val':
            xform = transforms.Compose([
                transforms.Resize(256),
                transforms.CenterCrop(224),
                transforms.ToTensor()])
        self.xform = xform



    def __len__(self):
        return len(self.imgs)



    def __getitem__(self, idxs):
        try:
            iter(idxs)
        except:
            idxs = idxs#[idxs]

        if torch.is_tensor(idxs):
            idxs = idxs.tolist()

        img_info = self.imgs[idxs]

        img_id = img_info['images'][0]['id']
        
        recipe_id = img_info['id']
        
        
        
        labels = ingredients_list_final_numpy_a[id_to_idx_dict[recipe_id]]
        
        pic = Image.open(self.partition+'/'+img_id[0]+'/'+img_id[1]+'/'+img_id[2]+'/'+img_id[3]+'/'+img_id)
        pic = self.xform(pic)
        return pic, labels

    def getinfo(self, ndx):
        return self.txts[ndx]

```

```python
from torchvision import datasets, models, transforms
from torch.autograd import Variable
import matplotlib.pyplot as plt

#Main training loop
use_gpu = torch.cuda.is_available()
print(use_gpu)

dataset1M_train = OneMRecipeDataset('train')
dataloader_train = torch.utils.data.DataLoader(dataset1M_train, batch_size=128, shuffle=True)

dataset1M_test = OneMRecipeDataset('test')
dataloader_test = torch.utils.data.DataLoader(dataset1M_test, batch_size=128, shuffle=True)

dataset1M_val = OneMRecipeDataset('val')
dataloader_val = torch.utils.data.DataLoader(dataset1M_val, batch_size=128, shuffle=True)

# vgg16 = models.vgg16(pretrained=True)


alexnetmodel = models.alexnet(pretrained=True)
for param in alexnetmodel.features.parameters():
    param.requires_grad = False

print(alexnetmodel)
    
features = list(alexnetmodel.classifier.children())[:-1]
alexnetmodel.classifier = nn.Sequential(*features)
if use_gpu:
    alexnetmodel.cuda()
    


```

```python
ingredient_layer = nn.Sequential(
    nn.Linear(4096,4096),
    nn.ReLU(inplace=True),
    nn.Linear(4096,833)
)

# feature_maps = torch.Tensor().cuda()
# ndxs = torch.Tensor().cuda()


optimizer = torch.optim.Adam(ingredient_layer.parameters(), lr=0.01)
EPOCHS = 5
losses = []
for epoch in range(EPOCHS):
    epoch_sum_loss = 0
    for i, data in enumerate(dataloader_train):
        print(i)
        inputs, labels = data
        alexnetmodel.train(False)
        alexnetmodel.eval()
        optimizer.zero_grad()
        alexnet_outputs = alexnetmodel(inputs)
        ingredients_output = ingredient_layer(alexnet_outputs)

        loss = nn.functional.binary_cross_entropy_with_logits(ingredients_output.double(), labels,pos_weight = torch.Tensor([100.]).double())
        print(loss)
        epoch_sum_loss += loss
        loss.backward()
        optimizer.step()
    losses.append(epoch_sum_loss)
```

```python
norm_losses = [1.3943]
for i in range(len(losses)):
    norm_losses.append(losses[i] / 545)
print(norm_losses)

t = np.arange(0,6)

fig, ax = plt.subplots()
ax.plot(t, norm_losses)

ax.set(xlabel='Epoch (s)', ylabel='Avg Binary Cross-Entropy Loss',
       title='Training loss for Image Ingredient Detection')
ax.set_ylim([0,1.5])

fig.savefig("test.png")
plt.show()

```
