```python
#convert dict of dict to list of playlists
def convert_string(input):
    converted_string = input
    if '(' in converted_string:
        string_index = converted_string.find('(')
        converted_string = converted_string[:string_index]
    if '-' in converted_string:
        string_index = converted_string.find('-')
        converted_string = converted_string[:string_index]
    punctuation = '''!()-[]{};:'"\,<>./?@# $%^&*_~'''
 
    for punctuation_mark in punctuation:
        if punctuation_mark in punctuation:
            converted_string = converted_string.replace(punctuation_mark, "")
    return converted_string.lower()


print(convert_string("Lose Control (feat. Ciara & Fat Man Scoop)"))

print(convert_string("U.N. Owen Was Her? (Remix)"))
```

```python
#convert dict of dict to list of playlists

converted_string_map_artists = {}
converted_string_map_tracks = {}


def update_string_map(data):

    playlist_list = []
    # print(data["playlists"][0]["tracks"])
    for playlist in data["playlists"]:
      # new_dictionary[] print(playlist["tracks"]
      playlist_songs = []
      for track in playlist["tracks"]:
        song = {}
        song["artist_name"] = track["artist_name"]
        song["track_name"] = track["track_name"]
        playlist_songs.append(song)
      playlist_list.append(playlist_songs)
    # print(playlist_list[0])

    for playlist in playlist_list:
        for song in playlist:
            converted_artist_name = convert_string(song["artist_name"])
            converted_track_name = convert_string(song["track_name"])
            if converted_artist_name not in converted_string_map_artists:
                converted_string_map_artists[converted_artist_name] = song["artist_name"]
                
            if converted_track_name not in converted_string_map_tracks:
                converted_string_map_tracks[converted_track_name] = song["track_name"]
```

```python
#convert dict of dict to list of playlists
def get_song_index(songkey):
    song_key_num = len(song_to_index)
    if songkey not in song_to_index:
        song_to_index[songkey] = song_key_num
        index_to_song[song_key_num] = songkey
    return song_to_index[songkey]
def update_dict(data):

    playlist_list = []
    # print(data["playlists"][0]["tracks"])
    for playlist in data["playlists"]:
      # new_dictionary[] print(playlist["tracks"]
      playlist_songs = []
      for track in playlist["tracks"]:
        song = {}
        song["artist_name"] = track["artist_name"]
        song["track_name"] = track["track_name"]
        playlist_songs.append(song)
      playlist_list.append(playlist_songs)
    # print(playlist_list[0])

    for playlist in playlist_list:
        for song in playlist:
            song_key = (song['artist_name'], song['track_name'])
            song_index = get_song_index(song_key)
            for song2 in playlist:
                song2_key = (song2['artist_name'], song2['track_name'])
                song2_index = get_song_index(song2_key)
                if song_index != song2_index:
                    if song_index in song_adjacency_dict:
                        song_adjacency = song_adjacency_dict[song_index]
                        if song2_index in song_adjacency:
                            song_adjacency_dict[song_index][song2_index] += 1
                        else:
                            song_adjacency[song2_index] = 1

                            song_adjacency_dict[song_index] = song_adjacency
                    else:
                        new_song_dict = {}
                        new_song_dict[song2_index] = 1
                        song_adjacency_dict[song_index] = new_song_dict
```

```python

with open('song_adjacency.txt','w') as file: 
    for song_index in range(len(song_to_index)):
        adjacent_song_list = []

        adjacent_songs = song_adjacency_dict[song_index]
        num_songs =  min(len(adjacent_songs), 25)
        sorted_songs = sorted(adjacent_songs, key=adjacent_songs.get, reverse=True)[:num_songs]
        adjacent_song_list.append(song_index)
        for i in range(num_songs):
            adjacent_song_list.append((sorted_songs[i], song_adjacency_dict[song_index][sorted_songs[i]]))

        file.write(str(adjacent_song_list) + '\n')
    file.close()
```

```python
#Code to open data files

index_to_song_from_file = {}
song_adjacency = {}
song_to_index_by_artist = {}
song_to_index_by_track = {}
song_to_index_by_both = {}

with open('song_lookup.txt','r') as file: 
    for line in file:
        song_info = line.rstrip()

        comma_index = song_info.find(',')
        comma_index2 = song_info[comma_index+2:].find(',') + comma_index + 2
        index = int(song_info[0:comma_index])
        artist = song_info[comma_index+2:comma_index2]
        track = song_info[comma_index2 + 2:]
        index_to_song_from_file[index] = (artist,track)
    file.close()

    
with open('song_adjacency.txt','r') as file: 
    for line in file:
        song_info = line.rstrip().split(", (")
        adjacency_list = []
        song_index = int(song_info[0][1:])
        for i in range(1,len(song_info)):
            token = song_info[i]
            comma_index = token.find(',')
            paren_index = token.find(')')
            value1 = int(token[0:comma_index])
            value2 = int(token[comma_index+2:paren_index])
            adjacency_list.append((value1,value2))
        song_adjacency[song_index] = adjacency_list
    file.close()

    


def get_index(item):
    if item[1] in song_adjacency:
        return song_adjacency[item[1]][0][1]
    else:
        return 1

with open('index_lookup_by_artist.txt','r') as file: 
    for line in file:
        song_info = line.rstrip()

        comma_index = song_info.find(',')
        comma_index2 = song_info.rfind(',')
        index = int(song_info[comma_index2+1:])
        artist = song_info[0:comma_index]
        track = song_info[comma_index + 2: comma_index2]
        if artist in song_to_index_by_artist:
            tracks = song_to_index_by_artist[artist]
            tracks.append((track,index))
            tracks = sorted(tracks,reverse=True,key=get_index)
            song_to_index_by_artist[artist] = tracks
        else:
            new_list = []
            new_list.append((track,index))
            song_to_index_by_artist[artist] = new_list
        if track in song_to_index_by_track:
            artists = song_to_index_by_track[track]
            artists.append((artist,index))
            artists = sorted(artists,reverse=True,key=get_index)
            song_to_index_by_track[track] = artists
        else:
            new_list = []
            new_list.append((artist,index))
            song_to_index_by_track[track] = new_list
        song_to_index_by_both[artist,track] = index
    file.close()
```

```python
#Given the artist and track, this returns a list of similar songs, along with the corresponding both-in-playlist count
def return_similar_songs(artist, track):
    similar_track_list = []
    if (artist,track) in song_to_index_by_both:
        song_index = song_to_index_by_both[(artist,track)]
        if song_index in song_adjacency:
            similar_song_list = song_adjacency[song_index]
            for (similar_track, similar_track_count) in similar_song_list:
                similar_track_artist, similar_track_track = index_to_song_from_file[similar_track]
                similar_track_list.append(similar_track_artist + ": " + similar_track_track + ", " + str(similar_track_count))
        return similar_track_list
    else:
        print("Artist + track combination not found")
```
