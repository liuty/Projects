```python
#convert dict of dict to list of playlists
def convert_string(input):
    converted_string = input
    if '(' in converted_string:
        string_index = converted_string.find('(')
        converted_string = converted_string[:string_index]
    if '-' in converted_string:
        string_index = converted_string.find('-')
        converted_string = converted_string[:string_index]
    punctuation = '''!()-[]{};:'"\,<>./?@# $%^&*_~'''
 
    for punctuation_mark in punctuation:
        if punctuation_mark in punctuation:
            converted_string = converted_string.replace(punctuation_mark, "")
    return converted_string.lower()


print(convert_string("Lose Control (feat. Ciara & Fat Man Scoop)"))

print(convert_string("U.N. Owen Was Her? (Remix)"))
```

```python
#convert dict of dict to list of playlists

converted_string_map_artists = {}
converted_string_map_tracks = {}


def update_string_map(data):

    playlist_list = []
    # print(data["playlists"][0]["tracks"])
    for playlist in data["playlists"]:
      # new_dictionary[] print(playlist["tracks"]
      playlist_songs = []
      for track in playlist["tracks"]:
        song = {}
        song["artist_name"] = track["artist_name"]
        song["track_name"] = track["track_name"]
        playlist_songs.append(song)
      playlist_list.append(playlist_songs)
    # print(playlist_list[0])

    for playlist in playlist_list:
        for song in playlist:
            converted_artist_name = convert_string(song["artist_name"])
            converted_track_name = convert_string(song["track_name"])
            if converted_artist_name not in converted_string_map_artists:
                converted_string_map_artists[converted_artist_name] = song["artist_name"]
                
            if converted_track_name not in converted_string_map_tracks:
                converted_string_map_tracks[converted_track_name] = song["track_name"]
```

```python
for i in range(len(playlist_file_names)):
# for i in range(1):
    filename = 'data/' + playlist_file_names[i]
    f = open(filename)
    print("Now processing {}".format(filename))
    data = json.load(f)
    update_string_map(data)

with open('string_map_tracks.pkl', 'wb') as handle:
    pickle.dump(converted_string_map_tracks, handle, protocol=pickle.HIGHEST_PROTOCOL)
with open('string_map_artists.pkl', 'wb') as handle:
    pickle.dump(converted_string_map_artists, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
